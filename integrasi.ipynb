```python
import numpy as np
import matplotlib.pyplot as plt

# mendefinisikan fungsi kuadrat yang akan diintegrasi
def f1(x):
    return x**2 - 4*x   # a1 = 1, b1 = -4, c1 = 0 ; 

def f2(x):
    return 4*x - x**2   # a2 = -1, b2 = 4, c2 = 0 ;

# dengan a, b, dan c adalah koefisien dari variabel-variabel fungsi kuadrat


# membuat array yang berisi nilai x
x = np.linspace(-5, 5, 100)

# menghitung nilai y untuk kedua kurva
y1 = f1(x)
y2 = f2(x)

# membuat plot untuk kedua kurva
plt.plot(x, y1, label='x¬≤ - 4x')
plt.plot(x, y2, label='4x - x¬≤')

# mengarsir luas wilayah di bawah kurva dari batas bawah samapai batas atas
plt.fill_between(x, y1, y2, where=(y1 <= y2), interpolate=True, alpha=0.25)

# menemukan titik potong menggunakan np.roots()
intersection_points = np.roots([2, -8, 0]) # a1 - a2 = 2 , b1 - b2 = -8, c1 - c2 = 0


# menampilkan titik potong pada plot
for point in intersection_points:
    plt.plot(point, f1(point), 'ro')

for point in intersection_points:
    plt.plot(point, f2(point), 'ro')

# beri judul plot dan label sumbu
plt.title('Intersection of x¬≤ - 4x and 4x - x¬≤')
plt.xlabel('x')
plt.ylabel('y')

# menampilkan koordinat kartesius
plt.grid()

# menampilkan legenda plot
plt.legend()

# menampilkan plot
plt.show()
```

```python
from scipy import integrate
import numpy as np

# mendefinisikan fungsi yang akan diintegrasikan
def f1(x):
    return (-x**2 + 4*x) # kurva atas

def f2(x):
    return (x**2 - 4*x) # kurva bawah


# substitusi batas atas dan bawah beserta subinterval

a = 0  # batas bawah
b = 4  # batas atas
N = 100  # jumlah subinterval atau partisi


x = np.linspace(a, b, N+1) # array yang berisi nilai x

# menghitung nilai y untuk kedua kurva
y1 = f1(x) 
y2 = f2(x)

# menghitung luas di bawah kurva dengan metode simpson
integral1 = integrate.simps(y1, x) # kurva atas
integral2 = integrate.simps(y2, x) # kurva bawah

# hasil akhirnya adalah mengurangi nilai kurva atas dengan kurva bawah
print("Hasil integral:", integral1 - integral2)
```

setelah mendapatkan nilai dari akar-akar polinomal, nilai tersebut dijadikan batas atas dan batas bawah untuk menghitung pengintegrasian numerik

```python
# definisikan fungsi yang akan diintegrasikan
def f(x):
    return x**3 - 3*x

# buat array x dari batas terkecil hingga batas terbesar dengan 1000 titik
x = np.linspace(-1.73205081, 1.73205081, 1000)

# menggambar grafik
plt.plot(x, f(x))

# mengisi area di bawah grafik dengan warna
plt.fill_between(x, f(x), where=(f(x) >= 0), interpolate = True, color='green', alpha=0.5)
plt.fill_between(x, f(x), where=(f(x) <= 0), interpolate = True, color='green', alpha=0.5)

# menambahkan koordinat kartesius pada gambar
plt.grid()

# menampilkan grafik
plt.show()



# perhitungan luas bidang datar di bawah grafik menggunakan metode trapesium :
# *pertama kali yang dilakukan adalah mengintegrasi kurva dari x1 sampai x2 dan kemudian dari x2 sampai x3

# dari x1 sampai x2 :
a, b = -1.73205081, 0
n = 1000
h = (b - a) / n
sum = (f(a) + f(b)) / 2.0

for i in range(1, n):
    x = a + i*h
    sum += f(x)

luas1 = h * sum

# dari x2 sampai x3 :
b, c = 0, 1.73205081
n = 1000
h = (c - b) / n
sum = (f(b) + f(c)) / 2.0

for i in range(1, n):
    x = b + i*h
    sum += f(x)

luas2 = h * sum

# hasil akhir dari pengintegrasian ini adalah menggabungkan nilai dari luas yang berada di atas sumbu-x (+) dengan luas yang berada di bawah sumbu-x (-)
print("luas bidang datar di bawah grafik adalah:", luas1 - luas2)
```
